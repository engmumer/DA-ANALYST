### Question 2:

How does time spent on social media compare to time spent on education among teens?

- Plot side-by-side histograms of Time on Social Media and Time on Education
- Find and compare the median time for both.
- Write simple comments to explain what the data shows

```python
bins=30
sns.histplot(df['Time_on_Social_Media'], color='blue', label='Social Media', bins=bins)
sns.histplot(df['Time_on_Education'], color='green', label='Education', bins=bins, alpha=0.5)

# Add labels and legend
plt.title('Time Spent on Social Media vs Education')
plt.xlabel('Hours per Day')
plt.ylabel('Number of People')
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
median_time_on_social_media=df['Time_on_Social_Media'].median()
print("Time spend on Social Media",median_time_on_social_media)
median_time_on_education=df['Time_on_Education'].median()
print("Time spend on education",median_time_on_education)
```

## Analysis
- The median time spent on Social Media is 2.5 hours per day.
- The median time spent on Education is 1.0 hour per day.
- This shows that, on average, people are spending more than twice as much time on social media compared to educational activities.
- It suggests a possible imbalance in daily time use, where entertainment may be prioritized over learning.
- This insight can be useful for awareness campaigns or to encourage more productive time habits.

### Question 3

- How can we detect and identify outliers in Daily Usage Hours, Exercise Hours, and Screen Time Before Bed among teens using the IQR method?
- Display the records corresponding to these outliers using boxplot.



```python
Q1_Daily_usage_hours=df['Daily_Usage_Hours'].quantile(0.25)
Q3_Daily_usage_hours=df['Daily_Usage_Hours'].quantile(0.75)
IQR_Daily_usage_hour=Q3_Daily_usage_hours-Q1_Daily_usage_hours
print(IQR_Daily_usage_hour)
```

```python
lower_bound_Daily_usage_hours = Q1_Daily_usage_hours - 1.5 * IQR_Daily_usage_hour
upper_bound_Daily_usage_hours = Q3_Daily_usage_hours + 1.5 * IQR_Daily_usage_hour
print(lower_bound_Daily_usage_hours)
print(upper_bound_Daily_usage_hours)
```

Above 10.265 and below -0.2374 the will be considered outlier 

```python
outliers1 = df[(df['Daily_Usage_Hours'] < lower_bound_Daily_usage_hours) |
              (df['Daily_Usage_Hours'] > upper_bound_Daily_usage_hours)]
outliers1
```

```python
Q1_Exe_hours=df['Exercise_Hours'].quantile(0.25)
Q3_Exe_hours=df['Exercise_Hours'].quantile(0.75)
IQR_Exe_hour=Q3_Exe_hours-Q1_Exe_hours
print(IQR_Exe_hour)
```

```python
lower_Exe_hours = Q1_Exe_hours - 1.5 * IQR_Exe_hour
upper_Exe_hours = Q3_Exe_hours + 1.5 * IQR_Exe_hour
print(lower_Exe_hours)
print(upper_Exe_hours)
```

```python
outliers2 = df[(df['Exercise_Hours'] < lower_Exe_hours) |
              (df['Exercise_Hours'] > upper_Exe_hours)]
outliers2
```

```python
Q1_Screen_time=df['Screen_Time_Before_Bed'].quantile(0.25)
Q3_Screen_time=df['Screen_Time_Before_Bed'].quantile(0.75)
IQR_Screen_time=Q3_Screen_time-Q1_Screen_time
print(IQR_Screen_time)
```

```python
lower_Screen_time = Q1_Screen_time - 1.5 * IQR_Screen_time
upper_Screen_time = Q3_Screen_time + 1.5 * IQR_Screen_time
print(lower_Screen_time)
print(upper_Screen_time)
```

```python
plt.figure(figsize=(10, 7))
sns.boxplot(data=df[['Daily_Usage_Hours', 'Exercise_Hours', 'Screen_Time_Before_Bed']], palette='GnBu_d')
plt.title('Outliers in Daily Usage, Exercise, and Screen Time', fontsize=14)
plt.ylabel('Hours')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Question 5

How does the average addiction level vary across different age groups? Show with line graph

```python
avg_addiction = df.groupby('Age')['Addiction_Level'].mean().reset_index()
plt.figure(figsize=(8, 5))
sns.lineplot(data=avg_addiction, x='Age', y='Addiction_Level', marker='o', linewidth=2)
plt.title('Average Addiction Level Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Average Addiction Level')
plt.grid(True)
plt.tight_layout()
plt.show()
```
