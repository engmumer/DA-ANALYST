```python
import sqlite3 as sql
import pandas as pd
```

```python
conn=sql.connect('database.sqlite')
conn
```

```python
def my_sql(q):
    return pd.read_sql_query(q,conn)
```

```python
q='''select * from salaries '''
my_sql(q)
```

1. What is the average TotalPayBenefits for each JobTitle? exclude ('Not Provided', 'Not provided')

```python
q='''select AVG(TotalPayBenefits),jobtitle from salaries 
where jobtitle not in ('Not Provided', 'Not provided')
group by jobtitle
order by totalpaybenefits DESC '''
my_sql(q)
```

2. Which JobTitle has the most employees in each year? exclude ('Not Provided', 'Not provided')

```python
q='''select JobTitle, Count(*) as total_count,year from salaries
    where JobTitle not in ('Not Provided', 'Not provided')
    group by year,Jobtitle  
    order by year, total_count desc'''
my_sql(q)
    
```

3. What is the total number of unique employees? exclude ('Not Provided', 'Not provided')

```python
q='''SELECT 
    COUNT(DISTINCT EmployeeName) AS unique_employee_count
FROM 
    salaries
WHERE 
    EmployeeName NOT IN ('Not Provided', 'Not provided') '''
my_sql(q)
```

4. Show All Employees Ordered By Their TotalPayBenefits In Descending Order? exclude ('Not Provided', 'Not provided')

```python
q= '''select employeename,totalpaybenefits from salaries
    where employeename not in ('Not Provided', 'Not provided')
    order by
    totalpaybenefits desc '''
my_sql(q)
```

5. Show All Employees With A TotalPaybenefits Value Between 125,000 and 150,000 And A Job Title 'Firefighter'? exclude ('Not Provided', 'Not provided')

```python
q='''select employeename,totalpaybenefits,jobtitle from salaries where jobtitle=="Firefighter"
    and employeename not in  ('Not Provided', 'Not provided')
    and totalpaybenefits between 125000 and 150000 
    order by totalpaybenefits DESC'''
my_sql(q)
```

6. How many employees have no overtime pay or other pay? exclude ('Not Provided', 'Not provided')


```python
q='''select COUNT(Employeename) from salaries where employeename not in  ('Not Provided', 'Not provided')
    and overtimepay=0 
    and otherpay=0 '''
my_sql(q)
```

7. Show top 10 employees having (overtimepay+OtherPay) in bar plot?  exclude ('Not Provided', 'Not provided')

```python
q='''select (overtimepay+otherpay) as extra_pay,employeename from salaries
    where employeename not in ('Not Provided', 'Not provided')
    order by
    extra_pay desc limit 10 '''
my_sql(q)
```

```python
b = pd.read_sql_query(q,conn)
b.plot(kind='bar', x='EmployeeName' , y='extra_pay', figsize=(10,4),color='lightsalmon')
```

8. Show The Average of BasePay, OverTime, OtherPay for all employees in pie chart? exclude ('Not Provided', 'Not provided')


```python
q= '''SELECT 
    AVG(BasePay) as avg_basepay,
    AVG(OvertimePay) as avg_overtimepay,
    AVG(OtherPay) as avg_otherpay
FROM salaries
WHERE 
    EmployeeName not in ('Not Provided', 'Not provided') AND 
    BasePay not in ('Not Provided', 'Not provided') AND
    OvertimePay not in ('Not Provided', 'Not provided') AND
    OtherPay not in ('Not Provided', 'Not provided') '''
my_sql(q)
```

```python
b = pd.read_sql_query(q,conn)
pie_data = b.melt(var_name='Pay Type', value_name='Amount')

# Now plot
pie_data.plot(kind='pie', y='Amount', labels=pie_data['Pay Type'], autopct='%1.1f%%', legend=False)
```

9. Get the average TotalPayBenefits for each JobTitle, but only for those with an average pay greater than 75000? exclude ('Not Provided', 'Not provided')


```python
q= '''SELECT 
    JobTitle, 
    AVG(TotalPayBenefits) AS avg_totalpaybenefits
FROM salaries
WHERE 
    EmployeeName NOT IN ('Not Provided', 'not provided') AND
    JobTitle NOT IN ('Not Provided', 'not provided') AND
    TotalPayBenefits != 0
GROUP BY JobTitle
HAVING AVG(TotalPayBenefits) > 75000
ORDER BY avg_totalpaybenefits DESC '''
my_sql(q)
```

10. Get the number of employees (EmployeeName) for each JobTitle, where the count is greater than 10?  exclude ('Not Provided', 'Not provided')


```python
q='''SELECT 
    JobTitle, 
    COUNT(EmployeeName) AS employee_count
FROM salaries
WHERE 
    EmployeeName NOT IN ('Not Provided', 'not provided') AND
    JobTitle NOT IN ('Not Provided', 'not provided')
GROUP BY JobTitle
HAVING COUNT(EmployeeName) > 10
ORDER BY employee_count DESC '''
my_sql(q)
```
