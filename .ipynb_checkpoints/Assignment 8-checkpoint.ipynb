```python
import pandas as pd
import numpy as np 
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Cars Company.csv")
df
```









#### 1. Check and Drop Duplicate Records

```python
df[df.duplicated()]
```

```python
df=df.drop_duplicates()
df
```

```python
df.count()
```

#### 2. Check and Drop missing or null values

```python
df.isnull().sum()
```

```python
df=df.dropna()
df
```

```python
df.count()
```

#### 3. What are the top 10 average MSRP (Price) per car brand (rounded figure)? with Horizontal Bar Plot

```python
MSRP_AVG=df.groupby('Make')['MSRP'].mean().round()
MSRP_AVG
```

```python
top10_MSRP_AVG=MSRP_AVG.sort_values(ascending=False).head(10)
top10_MSRP_AVG
```

```python
top10_MSRP_AVG.plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.xlabel('Average MSRP')
plt.ylabel('Brand')
plt.title('Top 10 Car Brands by Average MSRP')
```

#### 4. How has the MSRP (price) changed over the years (rounded figure)? with Line Plot

```python
df.Year.unique()
```

```python
Price_change_per_year=df.groupby('Year')['MSRP'].mean().round()
Price_change_per_year.plot(kind='line', figsize=(12, 6), color='Yellow')
plt.title('Average MSRP Over the Years')
plt.xlabel('Year')
plt.ylabel('Average MSRP (Rounded)')
plt.grid(True)
plt.show()
```

#### 5. How Engine HP (power) and MPG (efficiency) have changed over the years? show with two line plots

Hint: Average of (city mpg + highway MPG)

```python
df['Total_MPG'] = df['city mpg'] + df['highway MPG']
```

```python
engine_hp_per_year = df.groupby('Year')['Engine HP'].mean().round()
total_mpg_per_year = df.groupby('Year')['Total_MPG'].mean().round()
```

```python
plt.plot(engine_hp_per_year.index, engine_hp_per_year.values, label='Average Engine HP', marker='o', color='tomato')
plt.plot(total_mpg_per_year.index, total_mpg_per_year.values, label='Average Total MPG (City + Highway)', marker='o', color='seagreen')
plt.title('Engine HP and MPG Over the Years')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

#### 6. What percentage of cars are Compact, Midsize, or Large? Show with Pie chart

```python

vehicle_counts = df['Vehicle Size'].value_counts()

# Step 2: Filter only Compact, Midsize, and Large
selected_sizes = vehicle_counts[['Compact', 'Midsize', 'Large']]
selected_total = selected_sizes.sum()
all_total = vehicle_counts.sum()

# Step 3: Calculate Percentage
percentage = (selected_total / all_total) * 100
print(f"Percentage of Compact, Midsize, or Large cars: {percentage:.2f}%")

# Step 4: Prepare Pie Chart
labels = selected_sizes.index
sizes = selected_sizes.values
colors = ['#66b3ff','#99ff99','#ffcc99']

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Distribution of Compact, Midsize, and Large Cars')
plt.axis('equal')
plt.show()
```

#### 7. Which vehicle styles are sold the most? show with vertical bar plot

```python
vehicle_style_counts = df['Vehicle Style'].value_counts()

# Step 2: Plot Vertical Bar Chart
plt.figure(figsize=(12, 6))
vehicle_style_counts.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Most Sold Vehicle Styles')
plt.xlabel('Vehicle Style')
plt.ylabel('Number of Cars Sold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#### 8. Which car brands have the most vehicles priced at $40,000 or more and with a popularity score of at least 1000?

```python
filtered_df = df[(df['MSRP'] >= 40000) & (df['Popularity'] >= 1000)]

# Step 2: Count vehicles per brand
brand_counts = filtered_df['Make'].value_counts()

# Step 3: Display the result
print("Car brands with most vehicles priced ≥ $40,000 and Popularity ≥ 1000:")
print(brand_counts)
```
