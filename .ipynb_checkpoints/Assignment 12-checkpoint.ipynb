```python
import sqlite3 as sql
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#connection
conn = sql.connect("chinook.db")
conn
```

```python
def sql_query(q):
    return pd.read_sql_query(q,conn)
```

## Q1. Write a query to list the top 10 genres with the highest number of tracks. (Show in bar plot)

```python
q='''select count(t.track_id) as "number of track", g.name as "top 10 generes" from 
    track t 
    inner join genre g
    on 
    t.genre_id=g.genre_id
    group by g.name
    order by 1 DESC limit 10 '''
sql_query(q)
```

```python
b = pd.read_sql_query(q,conn)
b.plot(kind='bar', x='top 10 generes' , y='number of track', figsize=(10,4),color='lightsalmon')
```

## Q2. Write a query to list the top 10 albums that contain the most tracks.

```python
q='''select count(t.track_id) as "most track", a.title as "top 10 album" from 
    track t 
    inner join album a
    on 
    t.album_id=a.album_id
    group by a.title
    order by 1 DESC limit 10 '''
sql_query(q)
```

## Q3. Write a query to find the top 10 customers who spent the most money.

```python
q='''select c.customer_id, c.first_name, c.last_name,sum(i.total) as "most spent money" from 
    customer c
    inner join invoice i
    on 
    c.customer_id=i.customer_id
    group by c.customer_id
    order by 4 DESC limit 10 '''
sql_query(q)
```

## Q4. Write a query to find the top 5 artists who generated the highest revenue from track sales.

```python
q='''select ar.name "Artist_name",sum(il.unit_price*il.quantity) "Highest Revenue from track sales" from 
    invoice_line il
    inner join track t
    on 
    il.track_id=t.track_id
    inner join album b
    on
    t.album_id=b.album_id
    inner join artist ar
    on
    b.artist_id=ar.artist_id
    group by ar.name
    order by 2 DESC limit 5 '''
sql_query(q)
```

## Q5. Countries by Average Revenue per Invoice

```python
q='''select c.country as countries,AVG(i.total) "Average Revenue per invoice" from customer c
    inner join invoice i
    on
    c.customer_id=i.customer_id 
    group by c.country
    order by 2 DESC'''
sql_query(q)
```

## Q6. Write a query to find the media type with number of tracks.


```python
q='''select count(t.track_id) as "number of tracks",m.name as "Media type" from 
    track t
    inner join media_type as m 
    on
    t.media_type_id=m.media_type_id
    group by m.name
    order by
    1 desc'''
sql_query(q)
```

## Q7. Which artists have released tracks in more than one genre?

```python
q='''select ar.name as "Artist Name",Count(g.genre_id) as "Genre Count" from
    track t
    inner join genre g
    on
    t.genre_id=g.genre_id
    inner join album b
    on
    t.album_id=b.album_id
    inner join artist ar
    on
    b.artist_id=ar.artist_id
    group by ar.name
    having 
    Count(g.genre_id)>1
    order by 2 DESC
     '''
   
sql_query(q)
```

## Q8. Write a query to list the top 10 longest tracks by duration.

```python
q= ''' select  t.name "track name",t.milliseconds "longest track" 
    from track t
    order by 2 desc limit 10'''
sql_query(q)
```

## Q9. Top 5 Revenue Share percentage by Artist (Pie Chart)

```python
q='''select ar.name "Artist_name",sum(il.unit_price*il.quantity) "Highest Revenue from track sales" from 
    invoice_line il
    inner join track t
    on 
    il.track_id=t.track_id
    inner join album b
    on
    t.album_id=b.album_id
    inner join artist ar
    on
    b.artist_id=ar.artist_id
    group by ar.name
    order by 2 DESC limit 5 '''
sql_query(q)
```

```python
b= pd.read_sql_query(q,conn)
print(b.columns)
```

```python
plt.figure(figsize=(8,8))
plt.pie(b['Highest Revenue from track sales'], 
        labels=b['Artist_name'], 
        autopct='%1.1f%%', 
        startangle=140)
plt.title('Top 5 Artists by Revenue from Track Sales')
plt.axis('equal')
plt.show()
```

## Q10. Revenue by Genre (Rock, Latin, Metal) in Each Country (Stacked Column Chart)

```python
q='''SELECT 
    c.country,
    g.name AS genre,
    SUM(il.unit_price * il.quantity) AS revenue
FROM 
    invoice_line il
JOIN invoice i ON il.invoice_id = i.invoice_id
JOIN customer c ON i.customer_id = c.customer_id
JOIN track t ON il.track_id = t.track_id
JOIN genre g ON t.genre_id = g.genre_id
WHERE 
    g.name IN ('Rock', 'Latin', 'Metal')
GROUP BY 
    c.country, g.name
ORDER BY
    c.country, revenue DESC '''
sql_query(q)
```

```python
b= pd.read_sql_query(q, conn)

pivot_df = b.pivot(index='country', columns='genre', values='revenue').fillna(0)

pivot_df.plot(kind='bar', stacked=True, figsize=(12,7))

plt.title('Revenue by Genre (Rock, Latin, Metal) in Each Country')
plt.xlabel('Country')
plt.ylabel('Revenue')
plt.legend(title='Genre')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
