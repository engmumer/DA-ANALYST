#### ATM SIMULATOR

Python ATM Project Summary
Your project simulates a basic ATM system with the following core functionalities:

1. Global Variables
user_pin: Stores the correct PIN to access the ATM.

balance: Stores the user's current account balance (initially set to 500).

2. Functions
check_balance()
Prints the current balance of the user.

deposit()
Allows the user to deposit money into their account.

Accepts input for deposit amount.

Validates that the amount is a positive integer.

Updates and prints the new balance after deposit.

Handles invalid inputs (non-integer or negative amounts) gracefully.

transfer_money()
Allows the user to transfer money to another account.

Accepts a 14-digit account number (validates length and digits).

Takes the transfer amount input and validates it.

Checks if the balance is sufficient for transfer.

Deducts the amount from balance and confirms the transfer.

Handles errors and invalid inputs gracefully.

withdraw()
Allows the user to withdraw money from their account.

Takes withdrawal amount input and validates it.

Ensures sufficient balance before withdrawal.

Deducts the amount from balance and prints the remaining balance.

Handles invalid input or insufficient balance.

atm_menu()
Displays the ATM menu with options:

Check Balance

Deposit

Transfer Money

Withdraw

Exit
Takes the user's choice and calls the corresponding function.
Includes input validation for menu options.

start_atm()
Starts the ATM program by:

Asking the user to enter their PIN.

Validates the PIN against the stored user_pin.

If correct, displays the ATM menu (atm_menu()), else denies access.

3. Error Handling & Input Validation
All monetary inputs are wrapped in try-except blocks to catch non-integer inputs (e.g., letters or special characters) and prompt appropriate error messages without crashing.

Account number is validated to be exactly 14 digits.

Amounts are validated to be positive numbers and checked against the current balance for withdrawals and transfers.

4. User Interaction Flow
User starts by entering their PIN.

If PIN is correct, user is shown the menu options repeatedly until they choose to exit.

User performs transactions with input validation and feedback on success or failure.

```python
user_pin="jawan@123"
balance=500
def check_balance():
    print(f"your current balance is :${balance}")

# function for deposit of money
def deposit():
    global balance
    try:
        amount=int(input("Enter amount to deposit: $"))
        if amount>0:
            balance+=amount
            print(f"{amount}: $ successfully deposit, new balance: ${balance}")
        else:
             print("invalid deposit amount. Must be greater than zero")
    except ValueError:
        print("Invalid input please enter a number")

# function for transfer money
def transfer_money():
    try:
        global balance
        account_number=input("enter account number")
        if account_number.isdigit() and len(account_number)==14:
             amount=int(input(f"enter the amount for transfer to {account_number}: $"))
             if amount>0 and amount<=balance:
                balance-=amount
                print(f"{amount}: $ transfer, new balance: ${balance}")
             elif amount>balance:
                print("insufficient balance")
             else:
                print("Invalid Amount")
        else:
             print("Invalid account number")
    except ValueError:
        print("Invalid input please enter a number")

def withdraw():
    global balance
    try:
      amount=int(input("enter the amount: $"))
      if amount>0 and amount<=balance:
            balance-=amount
            print(f"your remaining balance is: ${balance}")
      else:
            print("Invalid amount or insufficient balance.")
    except ValueError:
        print("Invalid input. Please enter a number.")

def atm_menu():
    while True:
        print("\n=== ATM Menu ===")
        print("1. Check Balance")
        print("2. Deposit")
        print("3. Transfer Money")
        print("4. withdraw")
        print("5. Exit")

        choice = input("Choose an option (1-4): ")

        if choice == "1":
            check_balance()
        elif choice == "2":
            deposit()
        elif choice == "3":
            transfer_money()
        elif choice== "4":
            withdraw()
        elif choice == "5":
            print("Thank you for using the ATM. Goodbye!")
            break
        else:
            print("Invalid option. Try again.")

# Start the program
def start_atm():
    print("Welcome to the Python ATM!")
    pin = input("Enter your PIN: ")
    if pin == user_pin:
        print("PIN accepted.")
        atm_menu()
    else:
        print("Incorrect PIN. Access denied.")

# Run the ATM
start_atm()
               


```
